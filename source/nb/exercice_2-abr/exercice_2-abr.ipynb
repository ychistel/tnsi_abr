```python
from arbre_binaire import Arbre, Noeud
from draw_arbre import afficher

def appartient(x,a):
    if a.est_vide():
        return False
    else:
        if x < a.racine.valeur:
            return appartient(x,a.fils_gauche())
        elif x > a.racine.valeur:
            return appartient(x,a.fils_droit())
        else:
            return True
        
def ajoute(x,a):
    if a.est_vide():
        a = Arbre(x)
    else:
        if x < a.racine.valeur:
            if a.racine.gauche is None:
                a.racine.gauche = Noeud(x)
            else:
                ajoute(x,a.fils_gauche())
        if x > a.racine.valeur:
            if a.racine.droit is None:
                a.racine.droit = Noeud(x)
            else:
                ajoute(x,a.fils_droit())
    return a
```

```python
a = Arbre(5)
ajoute(6,a)
ajoute(2,a)
ajoute(4,a)
ajoute(1,a)
ajoute(3,a)
afficher(a)
```

```python
L=[25,19,32,13,24,27,33,21,29]
b = Arbre()
for valeur in L:
    b=ajoute(valeur,b)
afficher(b)
```

```python
L= [11,15,7,13,9,17,3]
c = Arbre()
for valeur in L:
    c=ajoute(valeur,c)
afficher(c)
```

```python
def parcours_infixe(arbre,p):
    if not arbre.est_vide():
        parcours_infixe(arbre.fils_gauche(),p)
        p.append(arbre.racine.valeur)
        parcours_infixe(arbre.fils_droit(),p)
    return p
```

```python
# Transforme une liste en Arbre binaire de recherche

def liste_en_abr(liste:list)->Arbre:
    """transforme une liste en Arbre binaire de recherche"""
    c = Arbre()
    for valeur in liste:
        c=ajoute(valeur,c)
    return c

a4 = liste_en_abr([8,4,6,9,11,3])
afficher(a4)
```

```python
# Transforme arbre binaire de recherche en liste

def abr_en_liste(arbre,p):
    if not arbre.est_vide():
        parcours_infixe(arbre.fils_gauche(),p)
        p.append(arbre.racine.valeur)
        parcours_infixe(arbre.fils_droit(),p)
    return p

L = abr_en_liste(a4,[])
print(L)
```
