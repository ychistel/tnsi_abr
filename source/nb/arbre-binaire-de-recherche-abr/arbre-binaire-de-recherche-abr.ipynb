## Partie 3

1. Fonction de recherche d'une valeur dans un ABR.
2. Fonction d'insertion d'un élément dans un ABR.

```python
def appartient(x,a):
    if a.est_vide():
        return False
    else:
        if x < a.racine.valeur:
            return appartient(x,a.fils_gauche())
        elif x > a.racine.valeur:
            return appartient(x,a.fils_droit())
        else:
            return True
```

```python
appartient(2,a)
```

```python
def ajoute(x,a):
    if a.est_vide():
        a.racine = Noeud(x)
    else:
        if x < a.racine.valeur:
            if a.racine.gauche is None:
                a.racine.gauche = Noeud(x)
            else:
                ajoute(x,a.fils_gauche())
        if x > a.racine.valeur:
            if a.racine.droit is None:
                a.racine.droit = Noeud(x)
            else:
                ajoute(x,a.fils_droit())
    return a     
```

```python
a = Arbre()
print(a)
```

```python
a = ajoute(8,a)
a = ajoute(5,a)
a = ajoute(11,a)
a = ajoute(7,a)
a =ajoute(9,a)
a =ajoute(12,a)
print(a)
```

```python
def parcours_infixe(a,p=[]):
    if not a.est_vide():
        p=parcours_infixe(a.fils_gauche(),p)
        p.append(a.racine.valeur)
        p=parcours_infixe(a.fils_droit(),p)
    return p
```

```python
print(parcours_infixe(a))
```

```python
def liste_en_abr(liste):
    a=Arbre()
    for valeur in liste:
        a = ajoute(valeur,a)
    return a

a = liste_en_abr([11,7,15,3,9,13,17])
```

```python
a=Arbre()

def liste_en_abr(liste,a,b=a):
    if len(liste) > 0:
        x = liste[0]
        if a.est_vide():
            a.racine = Noeud(liste.pop(0))
            liste_en_abr(liste,a,b)
        else:
            if x < a.racine.valeur:
                if a.racine.gauche is None:
                    a.racine.gauche = Noeud(liste.pop(0))
                    # appel récursif avec l'arbre entier enregistré dans b
                    liste_en_abr(liste,b)
                else:
                    liste_en_abr(liste,a.fils_gauche())
            if x > a.racine.valeur:
                if a.racine.droit is None:
                    a.racine.droit = Noeud(liste.pop(0))
                    # appel récursif avec l'arbre entier enregistré dans b
                    liste_en_abr(liste,b)
                else:
                    liste_en_abr(liste,a.fils_droit())
            
        return a


a = liste_en_abr([11,7,15,3,9,13,17,4,2,12],a)
```
